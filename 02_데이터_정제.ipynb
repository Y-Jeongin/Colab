#### 결측 데이터를 처리하는 방법은 크게 두가지가 있다.

## 피쳐들 간의 상관관계 분석

```python
drinks.columns
```

```python
# 'beer_servings'와 'wine_servings'
# pearson
corr = drinks[['beer_servings', 'wine_servings']].corr(method = 'pearson')
print(corr)
```

```python
# 여러 피쳐들간의 상관관계 분석
cols = ['country', 'beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol', 'continent']
corr = drinks[cols].corr(method = 'pearson')
print(corr)
```

```python
print(type(corr))
```

- DataFrame(data, columns, index)

```python
import seaborn as sns
```

```python
# heatmap(히트맵) : 피쳐들간의 상관계수를 시각화할 때 가장 많이 사용하는 도구
# 상관 행렬을 시각화
cols_view = ['beer', 'spirit', 'wine', 'alcohol']
sns.set(font_scale = 1.5)
hm = sns.heatmap(
    corr.values,
    cbar = True,
    annot = True,
    square = True,
    fmt = '.2f',
    yticklabels = cols_view,
    xticklabels = cols_view,
    annot_kws = {'size':15}
)
plt.tight_layout()
plt.show()
```

#### 대륙별 분석
- apply, agg

```python
# 대륙별 beer_servings의 평균, 최소값, 최대값, 합계를 계산
result = drinks.groupby('continent').beer_servings.agg(['mean', 'min', 'max', 'sum'])
result
```

```python
# 전체 알코올 평균보다 더 많은 알코올을 섭취하는 대륙을 추출하시오.
## 전체 알코올 평균
alcohol_mean = drinks.total_litres_of_pure_alcohol.mean()

## 대륙별 알코올 평균
continent_mean = drinks.groupby('continent').total_litres_of_pure_alcohol.mean()

## 평균보다 더 많은 알코올을 섭취하는 대륙
result = continent_mean[continent_mean >= alcohol_mean]
print(result)
```

```python
# 평균 beer_serings이 가장 높은 대륙을 구하시오.
## idxmax(), idxmin()
beer_continent = drinks.groupby('continent').beer_servings.mean().idxmax()
beer_continent
```

```python
beer_mean = drinks.groupby('continent').beer_servings.mean()
beer_mean.sort_values(ascending = False).index[0]
```

```python
# 대륙별 beer_servings의 평균, 최소값, 최대값, 합계를 계산
result = drinks.groupby('continent').beer_servings.agg(['mean', 'min', 'max', 'sum'])
result
```

```python
means = result['mean'].tolist()
mins = result['min'].tolist()
maxs = result['max'].tolist()
sums = result['sum'].tolist()

n_groups = len(result.index)

index = np.arange(n_groups)

bar_width = 0.1
gragh1 = plt.bar(index, means, bar_width, color = 'r', label = 'Mean')
gragh2 = plt.bar(index+bar_width, mins, bar_width, color = 'g', label = 'Min')
gragh3 = plt.bar(index+bar_width * 2, maxs, bar_width, color = 'b', label = 'Max')
gragh4 = plt.bar(index+bar_width * 3, sums, bar_width, color = 'y', label = 'Sum')

plt.xticks(index, result.index.tolist())
plt.legend()
plt.show()
```

```python
# 전체 알코올 평균보다 더 많은 알코올을 섭취하는 대륙을 추출하시오.
## 전체 알코올 평균
alcohol_mean = drinks.total_litres_of_pure_alcohol.mean()

## 대륙별 알코올 평균
continent_mean = drinks.groupby('continent').total_litres_of_pure_alcohol.mean()

print(continent_mean)
```

```python
continents = continent_mean.index.tolist()
continents.append('mean')

x_pos = np.arange(len(continents))

alcohol = continent_mean.tolist()
alcohol.append(alcohol_mean)

bar_list = plt.bar(x_pos, alcohol, align = 'center', alpha = 0.5)
bar_list[len(continents)-1].set_color('r')

plt.xticks(x_pos, continents)
plt.ylabel('알코올 소비량 평균')
plt.title('대륙별 알코올 소비량')

plt.plot([0, 6], [alcohol_mean,alcohol_mean], 'b--')  # plt.plot([시작위치 = x축], [마지막위치 = y축], 모양)

plt.show()
```

```python
# 평균 beer_serings이 가장 높은 대륙을 구하시오.
## idxmax(), idxmin()
beer_continent = drinks.groupby('continent').beer_servings.mean().idxmax()
beer_continent
```
